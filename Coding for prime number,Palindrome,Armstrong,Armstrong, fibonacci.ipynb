```python
#Question Prime / Not Prime
num=int(input("enter a number"))
if num>1:
    for i in range(2,num):
        if num%i==0:
            print(f"The number entered is: {num} \nwhich is Not prime number")
            break
    else:
        print(f"The value entered is: {num} \nwhich is Prime Number")
else:
    print("Please enter the number greater than 1")
```

```python
#question Palindrome
num=int(input("Enter any number"))
n=num
rev=0
while num>0:
    rev=(rev*10)+num%10
    num=num//10
if rev==n:
    print(f"Number {n} is palindrome")
else:
    print(f"Number {n} is not palindrome")
```

```python
char="MADAM"
rev=char[::-1]
if char==rev:
    print("The String is palindrome")
else:
    print("The String is not palindrome")
```

```python
#Question Armstrong
n=int(input("Enter any number:"))
x=n
sum=0
while (n>0):
    sum=sum+(n%10)*(n%10)*(n%10)
    n=n//10
if x==sum:
    print(f"Number {x} is Armstrong")
else:
    print(f"Number {x} is not Armstrong")
```

```python
# fibonacci
num=int(input("Enter a number"))
a,b=0,1
print(f"The entered number is {num} and output is")
for i in range(num):
    print(a,end=" ")
    c=a+b
    a,b=b,c
```
